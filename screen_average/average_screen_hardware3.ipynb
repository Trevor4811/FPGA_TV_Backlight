```python
# Import the necessary libraries
from PIL import Image
import numpy as np
from numpy import asarray
import time
from pynq import allocate
from pynq import Overlay
 
 
# load the image and convert into
# numpy array
img = np.array(Image.open(r'Sample6.png'))
numpydata = asarray(img)

# Create loops to find average r, g, b
r_array = numpydata[:,:,0]
g_array = numpydata[:,:,1]
b_array = numpydata[:,:,2]

starttime = time.time()

ol = Overlay("/home/xilinx/pynq/overlays/average_screen2/design_1.bit")

dma = ol.axi_dma
dma_send = ol.axi_dma.sendchannel
dma_recv = ol.axi_dma.recvchannel
hls_ip = ol.example_0 

CONTROL_REGISTER = 0x0
hls_ip.write(CONTROL_REGISTER, 0x81) # 0x81 will set bit 0

data_size = r_array.size
input_buffer = allocate(shape=(255960,), dtype=np.uint32)

count = 0
for i in range(360):
    for j in range(711):
        input_buffer[count] = r_array[i][j]
        count += 1
dma_send.transfer(input_buffer)

output_buffer = allocate(shape=(1,), dtype=np.uint32)

dma_recv.transfer(output_buffer)

r_value = output_buffer[0]
print(r_value)
del input_buffer, output_buffer
# end red calculations

ol = Overlay("/home/xilinx/pynq/overlays/average_screen2/design_1.bit")

dma = ol.axi_dma
dma_send = ol.axi_dma.sendchannel
dma_recv = ol.axi_dma.recvchannel
hls_ip = ol.example_0 

CONTROL_REGISTER = 0x0
hls_ip.write(CONTROL_REGISTER, 0x81) # 0x81 will set bit 0

data_size = g_array.size
input_buffer = allocate(shape=(255960,), dtype=np.uint32)

count1 = 0
for i in range(360):
    for j in range(711):
        input_buffer[count1] = g_array[i][j]
        count1 += 1
        
dma_send.transfer(input_buffer)

output_buffer = allocate(shape=(1,), dtype=np.uint32)

dma_recv.transfer(output_buffer)

g_value = output_buffer[0]
del input_buffer, output_buffer
#end green calculations

ol = Overlay("/home/xilinx/pynq/overlays/average_screen2/design_1.bit")

dma = ol.axi_dma
dma_send = ol.axi_dma.sendchannel
dma_recv = ol.axi_dma.recvchannel
hls_ip = ol.example_0 

CONTROL_REGISTER = 0x0
hls_ip.write(CONTROL_REGISTER, 0x81) # 0x81 will set bit 0

data_size = b_array.size
input_buffer = allocate(shape=(255960,), dtype=np.uint32)

count2 = 0
for i in range(360):
    for j in range(711):
        input_buffer[count2] = b_array[i][j]
        count1 += 1
        
dma_send.transfer(input_buffer)

output_buffer = allocate(shape=(1,), dtype=np.uint32)

dma_recv.transfer(output_buffer)

b_value = output_buffer[0]
del input_buffer, output_buffer
#end blue calculations

endtime = time.time() - starttime
print(endtime)
print(r_value)
print(g_value)
print(b_value)
```
